```python
from google.colab import userdata
from pinecone import Pinecone, ServerlessSpec
pinecone_api_key = userdata.get('PINECONE_API_KEY')
pc = Pinecone(api_key=pinecone_api_key)
```

# create Index



```python
index_name = "rag-project"

pc.create_index(
    name=index_name,
    dimension=768,
    metric="cosine",
    spec=ServerlessSpec(cloud="aws", region="us-east-1"),

)
index = pc.Index(index_name)
```

```python
from uuid import uuid4

from langchain_core.documents import Document

document_1 = Document(
    page_content="My name is Hamza Sheikh and i am a full stack developer",
    metadata={"source": "tweet"},
)

document_2 = Document(
    page_content="The weather forecast for tomorrow is cloudy and overcast, with a high of 62 degrees.",
    metadata={"source": "news"},
)

document_3 = Document(
    page_content="Building an exciting new project with LangChain - come check it out!",
    metadata={"source": "tweet"},
)

document_4 = Document(
    page_content="Robbers broke into the city bank and stole $1 million in cash.",
    metadata={"source": "news"},
)

document_5 = Document(
    page_content="The acting and visuals were absolutely stunning—I couldn’t take my eyes off the screen",
    metadata={"source": "tweet"},
)

document_6 = Document(
    page_content="Is the new iPhone worth the price? Read this review to find out.",
    metadata={"source": "website"},
)

document_7 = Document(
    page_content="The top 10 soccer players in the world right now.",
    metadata={"source": "website"},
)

document_8 = Document(
    page_content="LangGraph is the best framework for building stateful, agentic applications!",
    metadata={"source": "tweet"},
)

document_9 = Document(
    page_content="I’m already recommending it to my friends, and I can’t wait to experience it again myself!",
    metadata={"source": "news"},
)

document_10 = Document(
    page_content="I have a bad feeling I am going to get deleted :(",
    metadata={"source": "tweet"},
)

documents = [
    document_1,
    document_2,
    document_3,
    document_4,
    document_5,
    document_6,
    document_7,
    document_8,
    document_9,
    document_10,
]
```

```python
uuids = [str(uuid4()) for _ in range(len(documents))] #generate unique Ids

vector_store.add_documents(documents=documents, ids=uuids)
```

```python
len(documents)
```

#Performing a simple similarity search can be done as follows:

```python
result = vector_store.similarity_search(
    "I was completely hooked from start to finish; the story was so captivating!",
    k=3,
    filter={"source": "tweet"},
)
for res in result:
  print(f"${res.page_content} [{res.metadata}]")
```

#Similarity search with score
You can also search with score:

```python
result = vector_store.similarity_search_with_score(
    "It’s rare to find a movie that keeps you thinking about it long after it’s over",
    k=2,
    filter={"source": "tweet"}

)

for res,score in result:
  print(f"$ [SIM={score:3f} {res.page_content}] [{res.metadata}]")
```

```python
from langchain_google_genai import ChatGoogleGenerativeAI
llm = ChatGoogleGenerativeAI(
    model="gemini-1.5-flash",
    temperature=0,
    max_tokens= None,
    min_tokens= None,
    max_retries=2,
)
```

```python
def ans_to_user(query:str):
  vector_result = vector_store.similarity_search_with_score(query,  k=2)

  # pass to model vector result + query
  final_answer = llm.invoke(f" answer this query is${query} here are some reference to answer{vector_result}")
  return final_answer




```

```python
answer = ans_to_user("will it be cold tomorrow")
print(answer)
```
